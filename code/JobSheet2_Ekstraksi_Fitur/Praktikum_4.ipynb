```python
from PIL import Image
```

```python
# Langkah 1 - Load Image

# buka gambar dan konversi ke mode RGB
img = Image.open('../../data/images/lenna_test_image.png').convert("RGB")

# tampilkan gambar
img.show() 

# atau bisa mneggunakan display
display(img)
```

```python
# Langkah 2 - Ekstrak Fitur

# ekstrak channel menjadi bagian red, green, dan blue
r, g, b = img.split()

# cek panjang ukuran channel red
print(len(r.histogram()))

# cetak histogram pada channel red
print(r.histogram())
```
